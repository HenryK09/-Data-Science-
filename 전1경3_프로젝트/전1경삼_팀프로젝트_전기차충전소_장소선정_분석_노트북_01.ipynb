```python
# 실행 결과를 저장할 데이터프레임 정의
df_final_dataset = df_instl_tgt.copy()

df_final_dataset['충전소수'] = 0
df_final_dataset['전기차량수'] = 0.0
df_final_dataset['최단상권거리'] = 0.0
df_final_dataset['최단상권유동인구'] = 0
df_final_dataset['사고다발지역수'] = 0
df_final_dataset['행정동인구수'] = 0
df_final_dataset['스코어'] = 0

df_final_dataset
```

```python
# 설치 예정지 목록을 기준으로 반복문 실행
for index, row in df_final_dataset.iterrows():
  
  # 설치 예정지 위경도 추출
  tgt_lat = row['위도']
  tgt_lon = row['경도']
  tat_hangdong = row['행정동']
  ##############################################################################
  # Step 1. 설치 예정지 <-> 주변 1키로 이내 충전소 수 계산
  ##############################################################################

  # 거리 비교 목록
  distance_list = []

  for index2, row2 in df_charger.iterrows():
    # 2. 관악구 충전소 목록위 위경도 추출
    chgr_lat = row2['위도']
    chgr_lon = row2['경도']

    start = (float(tgt_lat), float(tgt_lon)) # (lat, lon)
    goal = (float(chgr_lat), float(chgr_lon))
    #위와 같이 출발지, 도착지 두 곳의 위경도 값을 start와 goal 변수에 저장합니다. 순서는 위도, 경도 순서입니다.
    #※두 지점 거리 계산하기 (킬로미터)
    distance = haversine(start, goal, unit='km')
    distance_list.append(distance)

  # 측정된 거리 목록 출력
  # print(distance_list)
  # 측정된 거리 목록 출력 (정렬됨)
  #print(np.sort(distance_list))

  # 1km 이내 충전소 필터링
  inner_1km = list(filter(lambda x: x <= 1, distance_list))
  # print(inner_1km)
  # print(f'[{index}] 주변 1km 이내 전기차 충전소는 {len(inner_1km)} 개 입니다.')
  df_final_dataset.loc[index,'충전소수'] = len(inner_1km)

  ##############################################################################
  # Step 2. 주소지에 해당하는 동에 등록된 전기차량의 수
  ##############################################################################
  
  # 21.01예상 수치 기준으로 전기차 수 산정
  predict_e_car_cnt = df_elec_car.loc[tat_hangdong]['21.01예상']
  # print(f" {tat_hangdong}의 예상 전기차 수는 {predict_e_car_cnt} 입니다.")
  df_final_dataset.loc[index,'전기차량수'] = predict_e_car_cnt


  ##############################################################################
  # Step 3. 최단 거리의 상권 선정
  ##############################################################################
  
  minDistance = 9999999999
  df_min_comm_dstr = None

  for index3, row3 in df_comm_dstr.iterrows():
    # 2. 관악구 충전소 목록위 위경도 추출
    # print(type(row3))
    comm_lat = row3['위도']
    comm_lon = row3['경도']
    comm_Popul = row3['총_유동인구_수']

    start = (float(tgt_lat), float(tgt_lon)) # (lat, lon)
    goal = (float(comm_lat), float(comm_lon))
    #위와 같이 출발지, 도착지 두 곳의 위경도 값을 start와 goal 변수에 저장합니다. 순서는 위도, 경도 순서입니다.
    #※두 지점 거리 계산하기 (킬로미터)
    distance = haversine(start, goal, unit='km')
    # print(f"{index} : {tgt_lat} {tgt_lon}")
    # print(f"{row3['상권_코드_명']} 와의 거리 {distance} / {comm_lat} {comm_lon}")
    # print(distance)
    if distance <= minDistance:
      minDistance = distance
      df_min_comm_dstr = row3
      # print(df_min_comm_dstr['상권_코드_명'])
    
    # distance_list.append(distance)

  # print(f"{index} 에서 제일 가까운 상권은 [{df_min_comm_dstr['상권_코드_명']}] 입니다.")
  df_final_dataset.loc[index,'최단상권거리'] = minDistance


  ##############################################################################
  # Step 4. 최단거리의 상권의 추정 유동인구 전체
  ##############################################################################
  
  # print(f"{df_min_comm_dstr['상권_코드_명']} 상권의 유동인구는 {df_min_comm_dstr['총_유동인구_수']} 명 입니다.")
  df_final_dataset.loc[index,'최단상권유동인구'] = df_min_comm_dstr['총_유동인구_수']

  ##############################################################################
  # Step 5. 반경 1km 이내 사고다발지역 수 (교통량 대체 자료)
  ##############################################################################
  #df_accident_area
  
  distance_list = []

  for index5, row5 in df_accident_area.iterrows():
    # 2. 관악구 충전소 목록위 위경도 추출
    accd_lat = row5['위도']
    accd_lon = row5['경도']

    start = (float(tgt_lat), float(tgt_lon)) # (lat, lon)
    goal = (float(accd_lat), float(accd_lon))
    #위와 같이 출발지, 도착지 두 곳의 위경도 값을 start와 goal 변수에 저장합니다. 순서는 위도, 경도 순서입니다.
    #※두 지점 거리 계산하기 (킬로미터)
    distance = haversine(start, goal, unit='km')
    distance_list.append(distance)

  # 측정된 거리 목록 출력
  # print(distance_list)
  # 측정된 거리 목록 출력 (정렬됨)
  #print(np.sort(distance_list))

  # 1km 이내 충전소 필터링
  inner_1km_accd = list(filter(lambda x: x <= 1, distance_list))
  # print(inner_1km)
  # print(f'[{index}] 주변 1km 이내 사고다발지역은 {len(inner_1km_accd)} 개 입니다.')
  df_final_dataset.loc[index,'사고다발지역수'] = len(inner_1km_accd)


  ##############################################################################
  # Step 6. 주소지에 해당하는 동에 거주중인 전체 인구 수
  ##############################################################################
  #df_population_f

  # 21.01예상 수치 기준으로 전기차 수 산정
  # print(df_population_f)
  # print(f" !!!!!!!!!! {tat_hangdong}")
  # print(df_population_f[tat_hangdong])
  # print(df_population_f.loc[0, tat_hangdong])
  # # predict_population_cnt = df_population_f[tat_hangdong]
  # print(f" {tat_hangdong}의 거주중인 전체 인구 수는 {df_population_f.loc[0, tat_hangdong]} 입니다.")

  df_final_dataset.loc[index,'행정동인구수'] = {df_population_f.loc[0, tat_hangdong]}
```

# Phase 4. 모델 정의

1	2	3	4	5	단위
4	3	2	1	0	대

1000	800	600	400	200	m

4~	3	2	1	0	건
10000~	15000~	20000~	25000~	30000~	명

```python
# 모델 UDF 정의
def df_simulator_model(model_no, 충전소수, 전기차량수, 최단상권거리, 최단상권유동인구, 사고다발지역수, 행정동인구수):
  
  # 충전소수 (낮을수록 고점)
  if 충전소수 == 0:
    충전소수_점수 = 5
  elif 충전소수 == 1:
    충전소수_점수 = 4
  elif 충전소수 == 2:
    충전소수_점수 = 3
  elif 충전소수 == 3:
    충전소수_점수 = 2
  elif 충전소수 >= 4:
    충전소수_점수 = 1

  # 전기차량수 (높을수록 고점)
  if 전기차량수 >= 500:
    전기차량수_점수 = 5
  elif 전기차량수 >= 400:
    전기차량수_점수 = 4
  elif 전기차량수 >= 300:
    전기차량수_점수 = 3
  elif 전기차량수 >= 200:
    전기차량수_점수 = 2
  elif 전기차량수 < 200:
    전기차량수_점수 = 1

  # 최단상권거리 (낮을수록 고점)
  if 최단상권거리 <= 0.2:
    최단상권거리_점수 = 5
  elif 최단상권거리 <= 0.4:
    최단상권거리_점수 = 4
  elif 최단상권거리 <= 0.6:
    최단상권거리_점수 = 3
  elif 최단상권거리 <= 0.8:
    최단상권거리_점수 = 2
  elif 최단상권거리 > 0.8:
    최단상권거리_점수 = 1

  # 최단상권유동인구 (높을수록 고점)
  if 최단상권유동인구 >= 1000000:
    최단상권유동인구_점수 = 5
  elif 최단상권유동인구 >= 800000:
    최단상권유동인구_점수 = 4
  elif 최단상권유동인구  >= 600000:
    최단상권유동인구_점수 = 3
  elif 최단상권유동인구  >= 400000:
    최단상권유동인구_점수 = 2
  elif 최단상권유동인구  < 400000:
    최단상권유동인구_점수 = 1

  # 사고다발지역수 (낮을수록 고점)
  if 사고다발지역수 == 0:
    사고다발지역수_점수 = 5
  elif 사고다발지역수 == 1:
    사고다발지역수_점수 = 4
  elif 사고다발지역수 == 2:
    사고다발지역수_점수 = 3
  elif 사고다발지역수 == 3:
    사고다발지역수_점수 = 2
  elif 사고다발지역수 >= 4:
    사고다발지역수_점수 = 1

  # 행정동인구수 (높을수록 고점)
  if 행정동인구수 >= 30000:
    행정동인구수_점수 = 5
  elif 행정동인구수 >= 25000:
    행정동인구수_점수 = 4
  elif 행정동인구수 >= 20000:
    행정동인구수_점수 = 3
  elif 행정동인구수 >= 15000:
    행정동인구수_점수 = 2
  elif 행정동인구수 < 15000:
    행정동인구수_점수 = 1

  if model_no == 1:
    충전소수_배율         = 6
    전기차량수_배율       =	5
    최단상권거리_배율     = 1.4	
    최단상권유동인구_배율 = 2.6	
    사고다발지역수_배율   = 3	
    행정동인구수_배율     = 2
  if model_no == 2:
    충전소수_배율         = 6
    전기차량수_배율       =	5
    최단상권거리_배율     = 3	
    최단상권유동인구_배율 = 4	
    사고다발지역수_배율   = 1	
    행정동인구수_배율     = 1
  if model_no == 3:
    충전소수_배율         = 6
    전기차량수_배율       = 5
    최단상권거리_배율     = 4	
    최단상권유동인구_배율 = 3	
    사고다발지역수_배율   = 1	
    행정동인구수_배율     = 1

  print(f"충전소수_점수\t\t{충전소수_점수} x {충전소수_배율}\t\t=> {충전소수_점수 * 충전소수_배율}")
  print(f"전기차량수_점수 \t{전기차량수_점수} x {전기차량수_배율}\t\t=> {전기차량수_점수 * 전기차량수_배율}")
  print(f"최단상권거리_점수\t{최단상권거리_점수} x {최단상권거리_배율}\t\t=> {최단상권거리_점수 * 최단상권거리_배율}")
  print(f"최단상권유동인구_점수\t{최단상권유동인구_점수} x {최단상권유동인구_배율}\t\t=> {최단상권유동인구_점수 * 최단상권유동인구_배율}")
  print(f"사고다발지역수_점수\t{사고다발지역수_점수} x {사고다발지역수_배율}\t\t=> {사고다발지역수_점수 * 사고다발지역수_배율}")
  print(f"행정동인구수_점수\t{행정동인구수_점수} x {행정동인구수_배율}\t\t=> {행정동인구수_점수 * 행정동인구수_배율}")

  score = (충전소수_점수 * 충전소수_배율) + (전기차량수_점수 * 전기차량수_배율) + (최단상권거리_점수 * 최단상권거리_배율) + (최단상권유동인구_점수 * 최단상권유동인구_배율) + (사고다발지역수_점수 * 사고다발지역수_배율) + (행정동인구수_점수 * 행정동인구수_배율) 

  return score
```
